# 1절. 조건문(if)
```
if 조건:
    조건1이 참일 경우 수행할 명령어들
elif 조건2:
    조건1이 거짓이고, 조건2가 참일 경우 수행할 명령어들
else:
    모든 조건을 만족하지 않을 때 수행할 명령어들
```


````python
score = int(input('점수는 :'))
if score >= 90:
    print('참 잘했어요')
    print('다음에도 기대할꼐요')
elif score >=60:
    print('분발하세요')
    print('뭐 그래도 고생하셨어요')
else:
    print('재시험 응시하셔야 합니다.')
    print('화이팅') # 반드시 형식에 맞춰야 한다 그렇지 않으면 전혀 다른 의미임
    
    
````

## 2.2 while문
```

while 조건 :
    조건이 참일 경우 반복할 문장들
else:
    조건이 거짓일 경우 한번 실행할 문장(else 구문은 선택사항)

```

````python
# 1 ~ 100까지 누적합 출력
tot = 0
i = 1
while i <= 100:
    tot += i
    i += 1
print("1~100까지 합은", tot, i)
````

# 4절. 중첩루프 탈출

````python
# 중첩 루프 탈출하는 방법 1 : 플래그 변수 이용

flag = False 
for a in range(0, 3):  # a=0, 1, 2
    for b in range(1, 3):  # b = 1, 2
        if a==b:
            flag = True
            break # 반복문을 아예 빠져 나오기
        print(a, b)
    # break 시 빠져 나오는 부분
    if flag:
        break
    print('DONE')
````

````python
# 중첩 루프 탈출 방법 2 :  예외 발생 => 예외 처리
class BreakOutError(Exception):
    pass
try:
    for a in range(0,3): # a=0, 1, 2
        for b in range(1, 3): # b-1, 2
            if a==b:
                raise BreakOutError() # 중첩루프 빠져나가기
            print(a, b)
                
except BreakOutError:
    print('DONE')
````

````python
# 10. 1 ~ 100 사이의 정수를 맞추는 프로그램 구현 - 1단계
import random
 
 # (1) 1~100 사이의 난수 발생
randomNum = random.randint(1, 100)
print('컴퓨터가 발생한 임의의 수 :', randomNum) 
  # (2) 수를 입력받아 틀리면 UP/DOWN메시지를 출력, 맞추면 축하메시지와 break하는 반복문
while True:
    inputNum = int(input("1~100사이의 수를 맞춰 보세요 : "))
    if (inputNum < 1) | (inputNum > 100):
        print('수의 범위를 벗어났습니다.')
    elif randomNum < inputNum:
        print('DOWN')
    elif randomNum > inputNum:
        print('UP')
    else:
        print('축하 축하합니다. 맞췄어요')
        break
print('끝')
````

````python
# 10. 1 ~ 100 사이의 정수를 맞추는 프로그램 구현 - 2단계
# 입력된 숫자가 틀리면 min~max 사이의 수라고 알려주고 숫자를 맞추면 축하 메시지와 함께 종료합니다 맞출 때까지 반복

import random
 # (1) 1~100 사이의 난수 발생
ran_num = random.randint(1, 100)
min_ = 1
max_ = 100
while True:
    guess_num = int(input(f"{min_} ~ {max_} 사이의 숫자를 입력하세요 : "))
    if guess_num < ran_num:
        print("업업업")
        min_ = max(min_, guess_num + 1)
    elif guess_num > ran_num:
        print("다운다운")
        max_ = min(max_, guess_num - 1)
    else:
        print(f"드디어 정답이네요! ㅊㅋㅊㅋ(정답은 {ran_num}입니다.)")
        break
        

````

````python
# 10. 1 ~ 100 사이의 정수를 맞추는 프로그램 구현 - 3단계
# 입력된 숫자가 틀리면 min~max 사이의 수라고 알려주고 숫자를 맞추면 축하 메시지와 함께 종료합니다 맞출 때까지 반복

import random

# (1) 1~100 사이의 난수 발생
ran_num = random.randint(1, 100)
min_ = 1
max_ = 100
chance = 5

for count in range(1, chance + 1):
    guess_num = int(input(f"[{count}/5] {min_} ~ {max_} 사이의 숫자 입력: "))
    
    if guess_num < ran_num:
        print("업업!!")
        min_ = max(min_, guess_num + 1)
    elif guess_num > ran_num:
        print("다운다운!")
        max_ = min(max_, guess_num - 1)
    else:
        print(f"오호~ 정답 ㅊㅋㅊㅋ (정답은 {ran_num}입니다.)")
````

````python
# 10번 3단계
'''
1~100 사이의 정수를 맞추기 프로그램
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 숫자를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메세지 뿌리고 끝남
2단계 : 숫자를 못 맞추면 fRom~to를 조정, 맞추면 축하메세지 뿌리고 끝남
3단계 : 2단계의 기회를 5회 제한. 5회의 기회가 지나면 실패 메세지 뿌리고 끝남
(2단계와 3단계 연습문제)
'''
import random
# (1) 1~100사이의 난수 발생
fRom, to = 1, 100
inputNum = 0 # 사용자가 맞출 수가 저장될 변수
randomNum = random.randint(fRom, to)
print('컴퓨터가 발생한 임의의 수 :', randomNum)
message = '기회를 모두 사용하셨습니다. 실패하셨습니다'
# (2) 수를 입력받아 틀리면 범위 조정과 UP/DOWN 메세지 출력, 
#    맞추면 축하축하 메세지와 break하는 반복문
for cnt in range(5, 0, -1):
    inputNum = int(input(f"{fRom}~{to}사이의 수를 맞춰 보세요(남은기회 {cnt}회) :"))
    print(f"{fRom}~{to}사이의 {inputNum}입력")
    if (inputNum<fRom) | (inputNum>to):
        print('수의 범위를 벗어났습니다')
    elif randomNum < inputNum:
        print('DOWN ↓')
        to = inputNum - 1
    elif randomNum > inputNum:
        print('UP ↑')
        fRom = inputNum+1
    else:
        message = '축하 축하 맞추셨습니다 {}번만에 성공하셨습니다.'.format(6-cnt)
        break
print(message)
````
